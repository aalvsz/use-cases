# VisDrone Fine-Tune — End-to-End Pipeline

Fine-tune LibreYOLO9 on the VisDrone2019-DET aerial-imagery benchmark. Runs top-to-bottom on a free Colab GPU (T4). Expect ~30-45 min for the full run.

What each cell does:
1. Install LibreYOLO + dependencies.
2. Download VisDrone via HuggingFace, convert to YOLO format.
3. Fine-tune (default: 20 epochs, YOLOv9-s, imgsz 640).
4. Evaluate on the val split.
5. Inference on a sample image.
